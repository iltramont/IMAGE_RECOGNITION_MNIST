```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
```

# Load data

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

# Data inspection

```python
n_rows = 3
n_columns = 4
start_index = 0

fig, axes = plt.subplots(n_rows, n_columns, figsize=(16, 9))

for i in  range(n_rows*n_columns):
    ax = axes[i // n_columns, i % n_columns]
    image = x_train[start_index + i]
    label = y_train[start_index + i]
    ax.imshow(image, cmap="gray_r")
    ax.set_title("Label: {}".format(label))

plt.suptitle("Some x_train sample images with corresponding labels", fontsize="xx-large")
plt.tight_layout()
plt.show()
```

```python
series_train = pd.Series(y_train)
series_test = pd.Series(y_test)

palette = sns.color_palette("hls", 2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 9), tight_layout=True)

for i in range(10):
    rect_train = ax1.bar(i, series_train.value_counts()[i], color=palette[0])
    ax1.bar_label(rect_train)
    rect_test = ax2.bar(i, series_test.value_counts()[i], color=palette[1])
    ax2.bar_label(rect_test)

for ax in (ax1, ax2):
    ax.set_xlabel("Label")
    ax.set_ylabel("Count")
    ax.set_xticks(list(range(10)))

ax1.set_title("y_train labels distribution", fontsize="xx-large")
ax2.set_title("y_test labels distribution", fontsize="xx-large")

plt.show()
```
